```python
# Εισαγωγή βιβλιοθηκών
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.utils import class_weight
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, LSTM, Flatten
from tensorflow.keras.utils import to_categorical
from keras.callbacks import EarlyStopping, ModelCheckpoint

# Φόρτωση και προεπεξεργασία δεδομένων
def loadDataset(filename="pcap_data.csv"):
    # Διαβάζουμε το αρχείο csv
    df = pd.read_csv(filename)
    # Χαρακτηριστικά (features): όλες οι στήλες εκτός από το target
    X = df.drop(columns=["target"]).values
    # Ετικέτες (labels): η στήλη target
    y = df["target"].values.astype(int)

    # Κανονικοποίηση Min-Max
    scaler = MinMaxScaler()
    X = scaler.fit_transform(X)

    # Χωρισμός σε Train/Test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )
    # Χωρισμός Train σε Train/Validation
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.125, stratify=y_train, random_state=42
    )

    return X_train.astype("float32"), y_train.astype("int32"), \
           X_val.astype("float32"), y_val.astype("int32"), \
           X_test.astype("float32"), y_test.astype("int32")

# Καλεσμα της συνάρτησης
X_train, y_train, X_val, y_val, X_test, y_test = loadDataset()
nclass = len(np.unique(y_train))

# One-hot encoding για τις ετικέτες
y_train_cat = to_categorical(y_train, num_classes=nclass)
y_val_cat   = to_categorical(y_val,   num_classes=nclass)
y_test_cat  = to_categorical(y_test,  num_classes=nclass)

# Ορισμός μοντέλων
def model_dense(inshape, nclass, N=64, lr=1e-4):
    # Απλό πλήρως συνδεδεμένο νευρωνικό δίκτυο
    model = Sequential([
        Dense(N, activation="relu", input_shape=(inshape,)),
        Dense(N, activation="relu"),
        Dense(nclass, activation="softmax")
    ])
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model

def model_conv1D(inshape, nclass, N=64, lr=1e-4):
    # Μοντέλο με Conv1D (για ακολουθιακά δεδομένα)
    model = Sequential([
        Conv1D(N, 3, activation="relu", input_shape=(inshape,1)),
        Flatten(),
        Dense(nclass, activation="softmax")
    ])
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model

def model_lstm(inshape, nclass, N=64, lr=1e-4):
    # Μοντέλο με LSTM (για ακολουθιακά δεδομένα)
    model = Sequential([
        LSTM(N, input_shape=(inshape,1)),
        Dense(nclass, activation="softmax")
    ])
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    return model

# Επιλογή μοντέλου
inshape = X_train.shape[1]
model = model_dense(inshape=inshape, nclass=nclass)

# Υπολογισμός κλάσεων (για ισορροπία σε ανισόρροπα δεδομένα)
cw = class_weight.compute_class_weight(class_weight="balanced",
                                       classes=np.unique(y_train),
                                       y=y_train)
cw_dict = {i: w for i, w in enumerate(cw)}

# Callbacks (EarlyStopping και ModelCheckpoint)
earlyStopping = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
modelCheckPoint = ModelCheckpoint("best_model.h5", save_best_only=True, monitor="val_accuracy", mode="max")

# Εκπαίδευση μοντέλου
history = model.fit(X_train, y_train_cat,
                    validation_data=(X_val, y_val_cat),
                    epochs=50,
                    batch_size=32,
                    class_weight=cw_dict,
                    callbacks=[earlyStopping, modelCheckPoint],
                    verbose=1)

# Αξιολόγηση στο Test set
pred_probs = model.predict(X_test)
pred_y = pred_probs.argmax(axis=-1)

print("Αναφορά ταξινόμησης:\n", classification_report(y_test, pred_y))
cm = confusion_matrix(y_test, pred_y)
ConfusionMatrixDisplay(cm).plot(xticks_rotation=45)
plt.show()

# Γραφήματα εκπαίδευσης
plt.figure()
plt.plot(history.history['accuracy'], label="Ακρίβεια Εκπαίδευσης")
plt.plot(history.history['val_accuracy'], label="Ακρίβεια Validation")
plt.legend(); plt.title("Ακρίβεια"); plt.show()

plt.figure()
plt.plot(history.history['loss'], label="Απώλεια Εκπαίδευσης")
plt.plot(history.history['val_loss'], label="Απώλεια Validation")
plt.legend(); plt.title("Απώλεια"); plt.show()
```
